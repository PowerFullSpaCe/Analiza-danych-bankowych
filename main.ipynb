## 2.2 Binning
We are going to bin "age" variable

```python

bins = [0,18,30,65,150]
#bins = np.linspace(min(df["age"]),max(df["age"]),5)
group_names = ["Under 18","18-30","30-65","65+"]

df["age-binned"] = pd.cut(df["age"], bins, labels=group_names, include_lowest=True)
df[["age", "age-binned"]].head()
df["age-binned"].value_counts(sort=False)

```

```python
#creating a histogram to visualize age groups
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["age-binned"].value_counts(sort=False))

plt.pyplot.xlabel("age")
plt.pyplot.ylabel("count")
plt.pyplot.title("Age bins")

```
